```python
import cv2
import matplotlib.pyplot as plt
```

## Part 1: Image Manipulation with OpenCV

```python
img = cv2.imread('logo.png')

plt.figure()
# OpenCV image channel is BGR so we flip the channels to RGB
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
img = cv2.imread('logo.png')

## Parameters for Textbox 
text = "Prelab 7"
font = cv2.FONT_HERSHEY_SIMPLEX
fontscale = 1
color = (0,0,255)
thickness = 2

text_box = cv2.getTextSize(text, font, fontscale, thickness)
orig = ( int(img.shape[1]/2 - text_box[0][0]/2), 30)

cv2.putText(img, "Prelab 7", orig, font, fontscale, color, thickness, cv2.LINE_AA )

# Parameters for Rectangle
vertex1 = (int(img.shape[1]/2-10), int(img.shape[0]/2-10)  )
vertex2 = (int(img.shape[1]/2+10), int(img.shape[0]/2+10)  )
color = (255,0,0)

cv2.rectangle(img, vertex1, vertex2, color, 2 ) 

plt.figure()
# OpenCV image channel is BGR so we flip the channels to RGB
plt.imshow(img[:,:,::-1])
plt.show()
```
